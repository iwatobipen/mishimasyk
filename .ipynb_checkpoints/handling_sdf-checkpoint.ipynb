```python
from rdkit.Chem import PandasTools
import pandas as pd
```

```python
mol_df = PandasTools.LoadSDF("cdk2.sdf")
```

```python
mol_df.head(5)
```

```python
from rdkit.Chem import Descriptors
mol_df["MW"] = mol_df.ROMol.map(Descriptors.MolWt)
```

```python
print len(mol_df)
print mol_df.MW.mean()
```

```python
print len(mol_df[mol_df.MW <= 350])
```

```python
PandasTools.AddMurckoToFrame(mol_df, MurckoCol="Murcko", Generic=True)
```

```python
mol_df.head(5)
```

```python
murckos = [Chem.MolFromSmiles(mol) for mol in mol_df.Murcko ]
Draw.MolsToGridImage(murckos[:5])
```

```python
PandasTools.SaveSMILESFromFrame(mol_df,"cdk2.smi", NamesCol="ID")
```
